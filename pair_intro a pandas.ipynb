```python
# ¿Qué país consiguió más de 200 medallas de oro?

paises_oro = df_medallas[df_medallas["Oro"] > 200]

paises_oro
```

```python
# ¿Qué país consiguió más de 200 medallas de oro y más de 1000 de plata?

paises_oro_plata = df_medallas[(df_medallas["Oro"] > 200) & (df_medallas["Plata"] > 1000)]

paises_oro_plata
```

```python
# ¿Qué país consiguió más de 200 medallas de oro o más de 700 de bronce?

paises_oro_bronce = df_medallas[(df_medallas["Oro"] > 200) | (df_medallas["Bronce"] > 700)]

paises_oro_bronce
```

```python
# SERIES
# 3 Crea dos listas: Una que se llame "valores" que tenga 7 números aleatorios. Otra que se llame "indices" que tenga 7 strings que se llamen como los días de la semana.

valores = pd.Series(range(1, 8))

print(valores)

indices = pd.Series(["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"])

print(indices)
```

```python
# A partir de estas dos listas crea una Serie donde la variable "valores" sean los valores de la Serie y la variable indices sean los índices de la Serie.

valores.index = indices

print(valores)
```

```python
# A partir de las dos listas anteriores crea un diccionario usando un bucle for. Para hacer esto tenemos el método zip que nos permite iterar por dos listas a la vez. Más información sobre este método aquí.

diccionario_vacio = {}

for valor, indice in zip(valores, indices):
    diccionario_vacio[valor] = indice

print(diccionario_vacio)

# Con zip recorremos dos listas y las unimos al mismo tiempo
```

```python
# Crea una Serie con el diccionario creado en el punto anterior. ¿Qué forma tiene esta última Serie?

diccionario = pd.Series(diccionario_vacio)

diccionario.head(7)

# Los índices son de tipo string
```
