```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', None)
```

```python
df_world_data1 = pd.read_csv("files/world-data-2023_part1.csv", index_col=0)
df_world_data1.head(2)
```

```python
df_world_data2 = pd.read_csv("files/world-data-2023_part2.csv", index_col=0)
df_world_data2.head(2)
```

```python
df_world_data1.info()
```

```python
df_world_data2.info()
```

Hemos visualizado ambos ficheros csv. 
Ambas tienen diferente número de columnas, pero mismo número de filas. 
Ambos ficheros tienen el común la columna Country y están en el mismo idioma. Vamos a realizar la unión a través de esta columna.

```python
df_world_data = pd.merge(df_world_data1,df_world_data2,left_on ="Country",right_on="country")
```

```python
df_world_data.head(2)
```

```python
round(df_world_data.isnull().sum()/df_world_data.shape[0]*100, 2)

# Realizamos el conteo de nulos. La columna con mayor % es Minimum wage
```

```python
df_world_data["Minimum wage"]

# Visualizamos la columna Minimum wage. Es el nivel de salario mínimo en moneda local. Entendemos que el % de nulos se debe a que muchos países no tienen un salario mínimo definido. No nos preocupa el % de nulos.
```

```python
round(df_world_data.duplicated().sum()/df_world_data.shape[0]*100, 2)

# No tenemos filas duplicadas
```

```python
# EJERCICIOS DE LIMPIEZA
# 1 Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.

df_world_data.drop(["country"], axis=1, inplace=True)
```

```python
df_world_data.head()

# Mostramos el DataFrame para verificar que se ha eliminado correctamente
```

2 Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

- No tengan espacios.
- Estén en minúscula.
- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".
- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.
- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df_world_data.columns = (df_world_data.columns
                         .str.lower()
                         .str.replace(r"\(p/km2\)|\(km2\)|[()%]", "", regex=True)
                         .str.replace("density\np/", "density")
                         .str.replace("\n", "")
                         .str.replace(":","")
                         .str.strip() # strip() elimina espacios sobrantes al inicio/final
                         .str.replace(" ", "_")) # Opcional: cambia espacios por guiones bajos
```

```python
df_world_data.head(1)
```

```python
# 3 La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

df_world_data[["latitude", "longitude"]] = df_world_data["coordinates"].str.split(",", expand=True)

df_world_data.drop(["coordinates"], axis= 1, inplace=True)
```
